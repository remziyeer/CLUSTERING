İlk olarak numpy, pandas, sklearn, matplotlib kütüphanelerini ekleyelim

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
 
import pandas as pd
import numpy as np
```

iris veri seti “sklearn” kütüphanesi içerisinde mevcut.


```python
iris = datasets.load_iris()
```

```python
x = pd.DataFrame(iris.data)
x.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']
 
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
```

Bu kısımda 2 grafik üretilmiştir. Iris (zambak) çiçeğinin çanak yaprak (Sepal) ve taç yaprak (Petal) kısımlarına ait uzunluk ve genişlik bilgileri grafikte mevcut.

```python
plt.figure(figsize=(14,7))
 
colormap = np.array(['purple', 'green', 'orange'])
 
plt.subplot(1, 2, 1)
plt.scatter(x.Sepal_Length, x.Sepal_Width, c=colormap[y.Targets], s=40)
plt.title('Sepal')
 
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Petal')
plt.show()
plt.close()
```

“iris” verisi için “k-means” modelini uyguladık. k parametresi yani küme sayısını 3 olarak seçtik.

```python
model = KMeans(n_clusters=3)
model.fit(x)
```

```python
plt.figure(figsize=(14,7))
 
colormap = np.array(['purple', 'green', 'orange'])
 
plt.subplot(1, 2, 1)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Gerçek Sınıflandırma')
 
plt.subplot(1, 2, 2)
plt.scatter(x.Petal_Length, x.Petal_Width, c=colormap[model.labels_], s=40)
plt.title('k-means Sınıflandırma')
```
